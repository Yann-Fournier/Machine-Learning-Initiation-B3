## **Phase CRISP-DM 2 : Compréhension des Données**

**Objectif :** Explorer le dataset `diamonds.csv` pour comprendre l'importance des différentes caractéristiques des diamants.

### Description des caractéristiques

1. **Carat** : Poids du diamant. Plus le carat est élevé, plus le diamant est en général précieux.
2. **Cut (Coupe)** : Qualité de la coupe, classée en cinq niveaux (Médiocre, Bon, Très Bon, Premium, Idéal).
3. **Color (Couleur)** : Mesure de la teinte, allant de D (incolore) à J (jaune). Les diamants incolores sont plus rares et chers.
4. **Clarity (Clarté)** : Mesure des inclusions (imperfections), allant de "Sans défaut interne" (IF) à "Inclus" (I1).
5. **Depth (Profondeur)** : Pourcentage de la profondeur totale, qui influe sur la réflexion de la lumière.
6. **Table** : Largeur du sommet du diamant, en pourcentage. Un bon ratio de table améliore l’éclat.
7. **Dimensions (x, y, z)** : Longueur, largeur et hauteur du diamant, en millimètres.

Dans le contexte des diamants, une **inclusion** désigne une impureté ou une imperfection à l'intérieur de la pierre. Ces inclusions peuvent prendre diverses formes, comme des petites bulles d'air, des fissures, ou des minéraux emprisonnés dans le diamant pendant sa formation. 

### Pourquoi les inclusions sont-elles importantes ?

1. **Clarté** : Les inclusions influencent directement la **clarté** d'un diamant, un des principaux critères de qualité avec le carat, la couleur et la coupe. Moins il y a d'inclusions, plus le diamant est clair et précieux.
2. **Valeur** : Les inclusions rendent le diamant moins rare et donc moins cher. Par exemple, un diamant sans défaut interne visible est plus précieux qu'un diamant avec plusieurs inclusions.
3. **Classification** : Les gemmologues classifient les diamants selon leur niveau d'inclusions :
   - **IF (Internally Flawless)** : Sans défaut interne, donc très rare.
   - **VVS (Very Very Slightly Included)** : Avec des inclusions extrêmement légères.
   - **VS (Very Slightly Included)** : Avec des inclusions légères mais visibles sous forte loupe.
   - **SI (Slightly Included)** : Inclusions facilement visibles sous loupe, parfois à l'œil nu.
   - **I (Included)** : Inclusions visibles à l'œil nu, qui peuvent affecter la transparence.

Les inclusions sont donc un facteur clé dans l’évaluation de la qualité et de la valeur d’un diamant.
